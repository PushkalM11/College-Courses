## Q1 - Gray Scale Point Operations 

###### 1 - Perform simple linear point operations for the following cases
- J = P.I
- J = I + L
- J = P.I + L

```python
def scale_and_add(I, P, L):
    J = np.ceil(P*I + L)
    J[J>255] = 255
    return J

def LPO1(img, P):
    return scale_and_add(img,P,0)

def LPO2(img, L):
    return scale_and_add(img,1,L)

def LPO3(img, P, L):
    return scale_and_add(img,P,L)
```

###### In each case, verify that the histogram of J is related to the histogram of I according to the relation discussed in class.

```python
def compute_histogram(image):
    image = np.ceil(image)
    intensity_counts = np.zeros(256)
    for i in range(256):
        intensity_counts[i] = np.sum(image==i)
    return intensity_counts

hist_I = compute_histogram(Voyager)

hist_1_J = compute_histogram(LPO1(Voyager, 2.5))
hist_2_J = compute_histogram(LPO2(Voyager, 40))
hist_3_J = compute_histogram(LPO3(Voyager, 2.5, 20))
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (15, 5), sharey = True)
intensities = np.arange(256)
plt.ylim(0,55000)

axes[0].plot(intensities, hist_I)
axes[0].plot(intensities, hist_1_J, alpha = 0.7)
axes[0].set_title('J = P.I')

axes[1].plot(intensities, hist_I)
axes[1].plot(intensities, hist_2_J, alpha = 0.7)
axes[1].set_title('J = I + L')

axes[2].plot(intensities, hist_I)
axes[2].plot(intensities, hist_3_J, alpha = 0.7)
axes[2].set_title('J = P.I + L')

fig.legend(['Histogram of I', 'Histogram of J'])
fig.suptitle('Histograms of I and J', font = 'Times New Roman', fontsize = 20)
fig.supxlabel('Intensities', fontsize = 15)
fig.supylabel('Number of pixels with given intensity', fontsize = 13)
plt.show()
```

Observations from the above histograms - 
- The histogram of J shifts to the right in all three cases, since P>1 and L>0.
- In case 2, the histogram remains intact except a right shift, because scaling is unity.
- In case 1 and 3, rapid variations are introduced in the histogram, due to scaling.
- There is a buildup at 255 due to these operations.

###### 2 - Full scale contrast stretch (FSCS). Verify your result by printing the min and max pixel values before and after applying FSCS. 

```python
def FSCS(img):
    
    min_I, max_I = np.min(img), np.max(img)
    k = 256
    P = (k-1)/(max_I - min_I)
    L = -1 * min_I * P
    
    return LPO3(img, P, L)

Voyager_FSCS = FSCS(Voyager)
print(f'The min and max intensities before FSCS : {np.min(Voyager)} and {np.max(Voyager)}')
print(f'The min and max intensities after FSCS  : {np.min(Voyager_FSCS)} and {np.max(Voyager_FSCS)}')
```

###### 3 - Log Magnitude Compression

```python
def log_compression(img):
    J = 1 + img
    R = img.shape[0]
    C = img.shape[1]
    J_plain = np.zeros((R,C))
    for r in range(R):
        for c in range(C):
            J_plain[r][c] = J[r][c]
            if J[r][c]==0:             # PIL is automatically converting 256 to 0. Hence doing this.
                J_plain[r][c] = 256
            
    J = np.log(J_plain)
    return FSCS(J)

Voyager_log_compressed = log_compression(Voyager)

NGC_log_compressed = log_compression(NGC)
```

######  4 - Gamma correction: J = $I^{\gamma}$ with $\gamma$ = 1.4. Clip pixel values greater than 255 to 255.

```python
def gamma_correction(img):
    J = np.power(img, 1.4)
    J[J>255] = 255
    return J

Voyager_gamma_corrected = gamma_correction(Voyager)

NGC_gamma_corrected = gamma_correction(NGC)
```

###### 5 - Histogram flattening. Plot the histogram of the image before and after applying histogram flattening.

```python
def cumulative_histogram(hist):
    CH = np.zeros(256)
    for i in range(256):
        CH[i] = np.sum(hist[:i+1])
        
    return CH

def flatten_histogram(img):
    
    img_hist = compute_histogram(img)
    normalised_hist = img_hist/(np.sum(img_hist))
    CDF = cumulative_histogram(normalised_hist)
        
    J = CDF[img]
    J = FSCS(J)
            
    return J,compute_histogram(J)

Voyager_flattened_Image, Voyager_flattened_hist = flatten_histogram(Voyager)

NGC_flattened_Image, NGC_flattened_hist = flatten_histogram(NGC)
```

## Q2 - Image Zooming

###### Write a program to zoom an image by 1.5 times along each axis using the following techniques discussed in class:
###### 1. Nearest Neighbor

```python
def NearestNeighborZoom(img):
    
    R = img.shape[0] 
    C = img.shape[1]
    
    R_new = int(np.ceil(R * 1.5))
    C_new = int(np.ceil(C * 1.5))
    zoomed = np.zeros((R_new, C_new)) 
    
    # distances between pixels in the fitted zoomed image.
    dist_x = (R_new - 1)/(R - 1)
    dist_y = (C_new - 1)/(C - 1)

    for r in range(R):
        for c in range(C):
            
            # The nearest neighbor of (r1,c1), (r1,c2), (r2,c1) and (r2,c2) is (r,c) 
            
            r1, r2 = int(np.floor(r * dist_x)), int(np.ceil(r * (R_new - 1) / (R - 1)))  
            c1, c2 = int(np.floor(c * dist_y)), int(np.ceil(c * dist_y))
            
            I = img[r][c]
            zoomed[r1][c1], zoomed[r1][c2], zoomed[r2][c1], zoomed[r2][c2] = I, I, I, I
            
    return zoomed
```

###### 2. Bilinear Interpolation

```python
def BilinearInterpolationZoom(img):
    
    R = img.shape[0]
    C = img.shape[1]
    
    R_new = int(np.ceil(R * 1.5))
    C_new = int(np.ceil(C * 1.5))
    zoomed = np.zeros((R_new, C_new))
    
    dist_x = (C_new - 1)/(C - 1)
    dist_y = (R_new - 1)/(R - 1)
    
    for r in range(R_new):
        for c in range(C_new):
            
            # The nearest neighbors of (r,c) are (r1,c1), (r1,c2), (r2,c1) and (r2,c2)
            
            r1, r2 = int(np.floor(r / dist_y)), int(np.ceil(r / dist_y))  
            c1, c2 = int(np.floor(c / dist_x)), int(np.ceil(c / dist_x))
            
            I1, I2, I3, I4 = img[r1][c1], img[r1][c2], img[r2][c1], img[r2][c2]
            
            i,j = r * 1/R_new, c * 1/C_new
            
            y1, y2 = r1 * 1/R, r2 * 1/R 
            x1, x2 = c1 * 1/C, c2 * 1/C
            
            if (c1 == c2 and r1 == r2):
                mat = np.array([[1]])
                Intensities = np.array([I1])
                X = np.array([1])
                
            elif (r1 == r2):
                mat = np.array([[1, x1], [1, x2]])
                Intensities = np.array([I1, I2])
                X = np.array([1, j])
                
            elif (c1 == c2):
                mat = np.array([[1, y1], [1, y2]])
                Intensities = np.array([I1, I3])
                X = np.array([1, i])
                
            else:
                mat = np.array([[1, x1, y1, x1*y1], [1, x2, y1, x2*y1], [1, x1, y2, x1*y2], [1, x2, y2, x2*y2]])
                Intensities = np.array([I1, I2, I3, I4])
                X = np.array([1, j, i, i*j])

            params = np.matmul(np.linalg.inv(mat), Intensities)
            
            zoomed[r][c] = np.dot(X, params)            
    
    return zoomed
```

- By observing the above images, we find no much quality difference between the two zooming techniques. This might be probably because the zoom factor is just 1.5.
- But, analytically we can definitely say that bilinear interpolation will produce better quality images because it scales the neighboring pixel values according to the smoothness assumption. Whereas in nearest neighbor, the neighboring values are simply replicated. Therefore, bilinear interpolation provides a smoother image output. 
- Computationally, bilinear interpolation is more intense because of a large number of matrix computations involved.
- Therefore, for lower values of zoom, we could settle with nearest neighbor technique, and for higher zoom values, we need to go with bilinear interpolation for better quality images.

###### Write a program that implements the DFT / IDFT pair using the following techniques.


###### 1 - From first principles i.e., as a summation over two dimensions.

```python
def DFT1(img):
    
    DFT = np.zeros(img.shape, dtype = complex)
    M = img.shape[0]
    N = img.shape[1]
    
    results = np.zeros(img.shape, dtype = complex)
    
    for v in range(N):
        for r in range(M):
            res = 0
            for c in range(N):
                res += img[r][c] * np.exp(-2j * np.pi * (v * c / N))
            results[r][v] = res
            
    for u in range(M):
        for v in range(N):
            res = 0
            for m in range(M):
                res += results[m][v] * np.exp(-2j * np.pi * (u * m / M))
            DFT[u][v] = res
        
    return DFT               
```

###### 2 - Matrix multiplication using the complex exponential matrix.

```python
def DFT2(img):
    
    # since M = N, both W_M and W_N matrices will be one and the same. Let it be W matrix.
    N = img.shape[1]
    
    def W_N_k_i(N,k,i):
        return np.exp(-2j * np.pi * k * i/ N)
    
    W = np.zeros((N, N), dtype = complex)
    
    for r in range(N):
        for c in range(N):
            W[r][c] = W_N_k_i(N, r, c)
            
    return np.matmul(W, np.matmul(img, W))  
    
```

###### 1 - Display the image, its DFT magnitude, phase, DFT magnitude subject to the logarithmic transformation.

```python
def get_DFT_magnitude(DFT_matrix):
    return np.abs(DFT_matrix)

def get_DFT_phase(DFT_matrix):
    return np.angle(DFT_matrix)

def log_transform(DFT_magnitude):
    return (1 + np.log(DFT_magnitude))
```

###### 2 - Generate the centered DFT using $(−1)^{i+j} I $ and display its magnitude.

```python
def centeredDFT(img):
    
    shifted_img = np.zeros(img.shape, dtype = complex)
    
    for r in range(img.shape[0]):
        for c in range(img.shape[1]):
            shifted_img[r][c] = ((-1)**(r+c)) * img[r][c]
            
    return DFT2(shifted_img)
```

###### 3 - Verify that -  IDFT [ DFT [ I ] ] = I. 

```python
def IDFT2(DFT):
    
    N = DFT.shape[1]
    
    def W_N_k_i(N,k,i):
        return np.exp(2j * np.pi * k * i/ N)
    
    W_inv = np.zeros((N, N), dtype = complex)
    
    for r in range(N):
        for c in range(N):
            W_inv[r][c] = W_N_k_i(N, r, c)
            
    return np.matmul(W_inv, np.matmul(DFT, W_inv)) / (N*N)  

def recreate_image(img):
    
    DFT = DFT2(img)
    recreated = np.abs(IDFT2(DFT))
    
    return recreated
```

###### 4 - Show the importance of phase by transposing the phase matrix and then synthesizing the image.

```python
def transpose_phase_and_synthesize(img):
    DFT = DFT2(img)
    mag = get_DFT_magnitude(DFT)
    phs = get_DFT_phase(DFT)
    
    phs_T = phs.T
    DFT = np.multiply(mag, np.exp(complex(0, 1) * phs_T))
    return np.abs(IDFT2(DFT))
```

```python
fig = plt.figure(figsize = (14,6))

rows = 2
cols = 3

for i,img in enumerate(New_Images):
    
    transposed_phase = transpose_phase_and_synthesize(img)
    Names = [' original', ' with phase transposed']
    
    imgs = [img, transposed_phase]
    
    for j in range(2):
        fig.add_subplot(rows,cols, 3*j + i+1)
        plt.imshow(imgs[j], cmap = 'gray')
        plt.xticks([])
        plt.yticks([])
        plt.title(names[i] + Names[j], fontsize=15)
        
fig.tight_layout()
```

###### Repeat by setting the phase to 0.

```python
def set_phase_to_zero(img):
    DFT = DFT2(img)
    mag = get_DFT_magnitude(DFT)
    
    phs = np.zeros((mag.shape))
    DFT = np.multiply(mag, np.exp(complex(0, 1) * phs))
    return np.abs(IDFT2(DFT))
```

```python
fig = plt.figure(figsize = (14,6))

rows = 2
cols = 3

for i,img in enumerate(New_Images):
    
    zero_phase = set_phase_to_zero(img)
    Names = [' original', ' with phase zero']
    
    imgs = [img, zero_phase]
    
    for j in range(2):
        fig.add_subplot(rows,cols, 3*j + i+1)
        plt.imshow(imgs[j], cmap = 'gray')
        plt.xticks([])
        plt.yticks([])
        plt.title(names[i] + Names[j], fontsize=15)
        
fig.tight_layout()
```

###### Repeat by adding noise.

```python
def add_noise_to_phase(img):
    DFT = DFT2(img)
    mag = get_DFT_magnitude(DFT)
    phs = get_DFT_phase(DFT)
    
    phs_noise = phs + np.random.normal(loc = 0, scale = 1, size = phs.shape)
    DFT = np.multiply(mag, np.exp(complex(0, 1) * phs_noise))
    return np.abs(IDFT2(DFT))
```

```python
fig = plt.figure(figsize = (14,6))

rows = 2
cols = 3

for i,img in enumerate(New_Images):
    
    phase_with_noise = add_noise_to_phase(img)
    Names = [' original', ' with added noise to phase']
    imgs = [img, phase_with_noise]
    
    for j in range(2):
        fig.add_subplot(rows,cols, 3*j + i+1)
        plt.imshow(imgs[j], cmap = 'gray')
        plt.xticks([])
        plt.yticks([])
        plt.title(names[i] + Names[j], fontsize=15)
        
fig.tight_layout()
```

It is evident from all the above images that the phase actually carries a lot of important image information and tampering with it changes the image to a great extent. We can particularly see that in the last example.
