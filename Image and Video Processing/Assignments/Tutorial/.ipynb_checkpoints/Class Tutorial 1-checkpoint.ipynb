```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
```

```python
image = img.imread('Baboon.tiff')
plt.imshow(image)
plt.show()
```

```python
red_stream = image[:,:,0]
green_stream = image[:,:,1]
blue_stream = image[:,:,2]

plt.title("Red Channel")
plt.imshow(red_stream, cmap = 'gray')
plt.show()
plt.title("Green Channel")
plt.imshow(green_stream, cmap = 'gray')
plt.show()
plt.title("Blue Channel")
plt.imshow(blue_stream, cmap = 'gray')
plt.show()
```

```python
print(f"Maximum in Red stream: {max(red_stream.flatten())}")
print(f"Maximum in Green stream: {max(green_stream.flatten())}")
print(f"Maximum in Blue stream: {max(blue_stream.flatten())}")
```

```python
def plot_graph(color_stream, name):
    lst = np.zeros(256)
    for i in np.array(color_stream).flatten():
        lst[i] += 1
    plt.title(name)
    plt.plot(np.arange(0, 256, 1), lst)
    plt.show()
```

```python
plot_graph(red_stream, "Red")
plot_graph(green_stream, "Green")
plot_graph(blue_stream, "Blue")
```

```python
print(f"Average of Red stream: {np.sum(red_stream) / len(red_stream.flatten())}")
print(f"Average of Green stream: {np.sum(green_stream) / len(green_stream.flatten())}")
print(f"Average of Blue stream: {np.sum(blue_stream) / len(blue_stream.flatten())}")
```

```python
def binarise_image(color_stream, threshold, name):
    binary_image = np.zeros(color_stream.shape)
    n, m = color_stream.shape
    for i in range(n):
        for j in range(m):
            if color_stream[i][j] >= threshold:
                binary_image[i][j] = 1
    plt.title(name)
    plt.imshow(binary_image, cmap = 'gray')
    plt.show()
    return binary_image
```

```python
red_binarised_image = binarise_image(red_stream, 200, "Red stream")
green_binarised_image = binarise_image(green_stream, 150, "Green stream")
blue_binarised_image = binarise_image(blue_stream, 200, "Blue stream")
```

```python
def dialate_image(binary_image, Window, name):
    l = len(Window)
    dialated_image = np.zeros(binary_image.shape)
    n, m = binary_image.shape
    for i in range(n):
        for j in range(m):
            for elem in range(l):
                wi = i + Window[elem][0]
                wj = j + Window[elem][1]
                if wi >= 0 and wi < n and wj >= 0 and wj < m:
                    if binary_image[wi][wj] == 1:
                        dialated_image[i][j] = 1
                        break
    plt.title(name)
    plt.imshow(dialated_image, cmap = 'gray')
    plt.show()
    return dialated_image
```

```python
Window = [[0, -1], [-1, 0], [0, 0], [0, 1], [1, 0]]

red_dialated_image = dialate_image(red_binarised_image, Window, "Red stream")
green_dialated_image = dialate_image(green_binarised_image, Window, "Green stream")
blue_dialated_image = dialate_image(blue_binarised_image, Window, "Blue stream")
```
