# Question 1 - Optimal Decorrelating Transform

## Vectorizing the image<a name = '4'></a>

```python
def vectorize_image(image, patch_shape):
    N, M = image.shape
    n, m = patch_shape
    x, y = int(N / n), int(M / m)
    vectorized_image = np.zeros([x * y, n * m])
    
    for i in range(x):
        for j in range(y):
            i1, j1 = n * i, m * j
            i2 = x * i
            vectorized_image[i2 + j, :] = image[i1 : i1 + n, j1 : j1 + m].flatten()
    
    return vectorized_image.T
```

## Computing covariance matrix<a name = '5'></a>

```python
def compute_covariance_matrix(X):
    X = np.array(X)
    return (X @ X.T) / X.shape[1]
```

## Making features zero mean<a name = '6'></a>

```python
def make_zero_mean(X):
    mean_along_cols = np.average(X, axis = 1).reshape([X.shape[0], 1])
    return mean_along_cols, (X - mean_along_cols)
```

## Reconstruct image with compression level<a name = '7'></a>

### Selecting relevant eigenvectors of correlation matrix<a name = '20'></a>

```python
def compute_eigen_vec(corr_matrix):
    eig_vals, eig_vecs = LA.eig(corr_matrix)
    sorted_indices = np.flip(np.argsort(eig_vals))
    return eig_vecs[:, sorted_indices]

def compute_transformation_matrix(eig_vecs, compression_level):
    N, M = eig_vecs.shape
    P = np.zeros([N, compression_level])
    P = eig_vecs[ : , : compression_level]
    return P @ P.T
```

### Transforming the image<a name = '21'></a>

```python
def transform_image(image, eig_vecs, compression_level):
    P = compute_transformation_matrix(eig_vecs, compression_level)
    Y = P @ image
    return Y
```

### Reconstruction from vectorized version<a name = '22'></a>

```python
def reconstruct_image(vectorized_image, patch_shape):
    vectorized_image = vectorized_image.T
    n, m = patch_shape
    N, M = vectorized_image.shape
    N_new = int(np.sqrt(N * M))
    n_patch, m_patch = int(N_new / n), int(N_new / m)
    new_image = np.zeros([N_new, N_new])
    for i in range(N):
        i1, j1 = int(i / n_patch) * n, (i % m_patch) * m
        new_image[i1 : i1 + n, j1 : j1 + m] = vectorized_image[i, : ].reshape(patch_shape)
    return new_image
```

### Decorrelation using PCA<a name = '23'></a>

```python
def decorrelate_using_pca(image, patch_shape, compression_levels):
    compressed_images = [image]
    
    vectorized_image = vectorize_image(image, patch_shape)
    mean_along_cols, zero_mean_data = make_zero_mean(vectorized_image)
    corr_matrix = compute_covariance_matrix(zero_mean_data)
    eigen_vecs = compute_eigen_vec(corr_matrix)
    
    for compression_level in compression_levels:
        tmp_image_zero_mean = transform_image(vectorized_image, eigen_vecs, compression_level)
        tmp_image = tmp_image_zero_mean + mean_along_cols
        compressed_images.append(reconstruct_image(tmp_image, [8, 8]))
    
    return compressed_images
```

## Testing for different compression levels<a name = '8'></a>

```python
compression_list = [48, 32, 16, 6, 2]
compressed_images_list = decorrelate_using_pca(lighthouse, [8, 8], compression_list)
```

```python
fig = plt.figure(figsize = (14, 10))

name = ["Original lighthouse image", "Compression Level: "]

for i in range(len(compression_list) + 1):
    plt.subplot(2, 3, i + 1)
    plt.imshow(compressed_images_list[i], cmap = 'gray')
    if i == 0:
        plt.title(name[i], size = 15)
    else:
        plt.title(name[1] + f"{compression_list[i - 1]}", size = 15)
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
plt.show()
```

PCA is quite effective in decorrelating the image, this is due to the fact that it is an almost lossless coding scheme wherein we only take the first few significant eigenvalues and the rest are discarded.<br> From the above images, for compression levels 48, 32 and 16 are quite similar to the original image which clearly illustrates the effectiveness of PCA in compression.

# Question 2 - Discrete Cosine Transform

## Dividing image into non-overlapping blocks<a name = '10'></a>

```python
def divide_image(image, block_shape):
    N, M = image.shape
    n, m = block_shape
    x, y = int(N / n), int(M / m)
    block_images = []
    
    for i in range(x):
        for j in range(y):
            i1, j1 = n * i, m * j
            block_images.append(image[i1 : i1 + m, j1 : j1 + m])
    
    return block_images
```

## Type-$\mathrm{II}$ DCT<a name = '11'></a>

```python
def C_N(u):
    if u == 0:
        return np.sqrt(0.5)
    return 1

def DCT_image(image):
    N, M = image.shape
    k = 4 / (N * M)
    
    dct_image = np.zeros([N, M], dtype = float)
    temp_result = np.zeros([N, M], dtype = float)
    
    for v in range(M):
        for i in range(N):
            for j in range(M):
                temp_result[i, v] += image[i, j] * np.cos(np.pi * (2 * j + 1) * v / (2 * M))
    
    for u in range(N):
        for v in range(M):
            for i in range(N):
                dct_image[u, v] += temp_result[i, v] * np.cos(np.pi * (2 * i + 1) * u / (2 * N))
            dct_image[u, v] *= k * C_N(u) * C_N(v)
            
    return dct_image
```

## Dividing with Q matrix and quantization<a name = '12'></a>

```python
def divide_with_Q(dct_image):
    if dct_image.shape != (8, 8):
        print("Error! Please check shape of your image")
        return 
    Q = np.array([[16, 11, 10, 16, 24, 40, 51, 61], 
                  [12, 12, 14, 19, 26, 58, 60, 55], 
                  [14, 13, 16, 24, 40, 57, 69, 56], 
                  [14, 17, 22, 29, 51, 87, 80, 62], 
                  [18, 22, 37, 56, 68, 109, 103, 77], 
                  [24, 35, 55, 64, 81, 104, 113, 92], 
                  [49, 64, 78, 87, 103, 121, 120, 101], 
                  [72, 92, 95, 98, 112, 100, 103, 99]], dtype = float)
    return dct_image / Q

def quantize_image(dct_image_div_Q):
    return np.array((dct_image_div_Q + 0.5), dtype = int)
```

## Multiplying with Q matrix<a name = '13'></a>

```python
def multiply_with_Q(dct_image_div_Q):
    if dct_image_div_Q.shape != (8, 8):
        print("Error! Please check shape of your image")
        return 
    Q = np.array([[16, 11, 10, 16, 24, 40, 51, 61], 
                  [12, 12, 14, 19, 26, 58, 60, 55], 
                  [14, 13, 16, 24, 40, 57, 69, 56], 
                  [14, 17, 22, 29, 51, 87, 80, 62], 
                  [18, 22, 37, 56, 68, 109, 103, 77], 
                  [24, 35, 55, 64, 81, 104, 113, 92], 
                  [49, 64, 78, 87, 103, 121, 120, 101], 
                  [72, 92, 95, 98, 112, 100, 103, 99]], dtype = float)
    return dct_image_div_Q * Q
```

## Type-$\mathrm{II}$ IDCT<a name = '14'></a>

```python
def IDCT_image(dct_image):
    N, M = dct_image.shape
    
    idct_image = np.zeros([N, M], dtype = float)
    temp_result = np.zeros([N, M], dtype = float)
        
    for j in range(M):
        for u in range(N):
            for v in range(M):
                temp_result[u, j] += dct_image[u, v] * C_N(v) * np.cos(np.pi * (2 * j + 1) * v / (2 * M))
    
    for i in range(N):
        for j in range(M):
            for u in range(N):
                idct_image[i, j] += temp_result[u, j] * C_N(u) * np.cos(np.pi * (2 * i + 1) * u / (2 * N))
    
    return idct_image
```

## Rearranging sub-blocks to get original image<a name = '15'></a>

```python
def rearrange_images(idct_image_patches):
    num = len(idct_image_patches)
    side = int(np.sqrt(num))
    n, m = idct_image_patches[0].shape
    N, M = side * n, side * m
    
    reconstructed_image = np.zeros([N, M])
    
    for i in range(side):
        i1 = i * n
        for j in range(side):
            j1 = j * m
            reconstructed_image[i1 : i1 + n, j1 : j1 + m] = idct_image_patches[i * side + j]
    
    return reconstructed_image
```

## Encoder<a name = '16'></a>

```python
def encode_image(image, block_shape):
    image_patch_list = divide_image(image, block_shape)
    encoded_patches = []
    
    for image_patch in image_patch_list:
        dct_image_patch = DCT_image(image_patch)
        dct_image_patch_div_Q = divide_with_Q(dct_image_patch)
        quant_patch = quantize_image(dct_image_patch_div_Q)
        encoded_patches.append(quant_patch)
    return encoded_patches
```

## Decoder<a name = '17'></a>

```python
def decode_image(encoded_patches):
    idct_image_patches = []
    
    for encoded_patch in encoded_patches:
        patch_mult_Q = multiply_with_Q(encoded_patch)
        idct_image_patch = IDCT_image(patch_mult_Q)
        idct_image_patches.append(idct_image_patch)
    
    return rearrange_images(idct_image_patches)
```

## Verifying sparsity of quantized DCT<a name = '18'></a>

```python
quant_dct_patches = encode_image(lighthouse, [8, 8])
```

```python
non_zero_count = 0
for dct_patch in quant_dct_patches:
    non_zero_count += np.count_nonzero(dct_patch)
n, m = quant_dct_patches[0].shape
total = n * m * len(quant_dct_patches)

print(f"Total number of non-zero elements: {non_zero_count}")
print(f"Percentage of non-zero entries: {100 * non_zero_count / total:.3f}%")
```

Clearly the quantized DCT matrix is sparse!

## Testing on image<a name = '19'></a>
