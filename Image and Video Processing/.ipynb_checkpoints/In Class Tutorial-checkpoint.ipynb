```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
```

```python
image = img.imread('Baboon.tiff')
plt.imshow(image)
plt.show()
```

```python
red_stream = image[:,:,0]
green_stream = image[:,:,1]
blue_stream = image[:,:,2]

plt.title("Red Channel")
plt.imshow(red_stream, cmap = 'gray')
plt.show()
plt.title("Green Channel")
plt.imshow(green_stream, cmap = 'gray')
plt.show()
plt.title("Blue Channel")
plt.imshow(blue_stream, cmap = 'gray')
plt.show()
```

```python
print(f"Maximum in Red stream: {max(red_stream[0])}")
print(f"Maximum in Green stream: {max(green_stream[0])}")
print(f"Maximum in Blue stream: {max(blue_stream[0])}")
```

```python
def plot_hist(color_stream, name):
    hist = np.zeros(256)
    for i in np.array(color_stream).flatten():
        hist[i] += 1
    plt.title(name)
    plt.plot(np.arange(1, 257, 1), hist)
    plt.show()
```

```python
plot_hist(red_stream, "Red")
plot_hist(green_stream, "Green")
plot_hist(blue_stream, "Blue")
```

```python
def binarise_image(color_stream, threshold, name):
    binary_image = np.zeros(color_stream.shape)
    n, m = color_stream.shape
    for i in range(n):
        for j in range(m):
            if color_stream[i][j] >= 200:
                binary_image[i][j] = 1
    plt.title(name)
    plt.imshow(binary_image, cmap = 'gray')
    plt.show()
    return binary_image
```

```python
red_binarised_image = binarise_image(red_stream, 200, "Red stream")
green_binarised_image = binarise_image(green_stream, 200, "Green stream")
blue_binarised_image = binarise_image(blue_stream, 200, "Blue stream")
```

```python
def dialate_image(binary_image, Window, name):
    l = len(Window)
    dialated_image = np.zeros(binary_image.shape)
    n, m = binary_image.shape
    for i in range(n):
        for j in range(m):
            value = 0
            for elem in range(l):
                wi = i + Window[elem][0]
                wj = j + Window[elem][1]
                if wi >= 0 and wi <= 511 and wj >= 0 and wj <= 511:
                    if binary_image[wi][wj] == 1:
                        value = 1
                        break
            dialated_image[i][j] = value
    plt.title(name)
    plt.imshow(dialated_image, cmap = 'gray')
    plt.show()
    return dialated_image
```

```python
Window = [[0, -1], [-1, 0], [0, 0], [0, 1], [1, 0]]
red_dialated_image = dialate_image(red_binarised_image, Window, "Red stream")
green_dialated_image = dialate_image(green_binarised_image, Window, "Green stream")
blue_dialated_image = dialate_image(blue_binarised_image, Window, "Blue stream")
```
