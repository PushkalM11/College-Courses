```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('dictionary.json')
data = json.load(file)
meanings = []
l = 0
not_allowed = ".,-()[]0123456789#!;\+*`&:$%/<=>@^"
for word in data:
    meaning = data[word].translate({ord(c): None for c in not_allowed}).lower()
    meaning = meaning.translate({ord(c): None for c in "\"\'"})
    meanings += meaning.split()
    l += 1
print("Number of words:", l)
```

```python
unique, index, count = np.unique(meanings, return_index = True, return_counts = True)
```

```python
order = np.flip(np.argsort(count))
unique = unique[order]
index = index[order]
count = count[order]
rank = 1 + np.arange(len(index))
```

```python
from scipy.optimize import minimize_scalar, curve_fit, minimize
from scipy.stats import norm, lognorm
from scipy.special import zeta, gamma
```

```python
def mle_estimate(word_counts, function):
    params = []
    if function == "first_function":
        mle = minimize_scalar(ll_first_function, bracket=(1 + 1e-10, 4), args=word_counts, method="brent")
        beta = mle.x
        alpha = 1 / (beta - 1)
        C = ((np.arange(len(word_counts)) + 1) ** (-alpha)).sum()
        params = {"alpha" : alpha, "C" : C}
        
    elif function == "second_function":
        mle = minimize_scalar(ll_second_function, bracket=(1 + 1e-10, 4), args=word_counts, method="brent")
        beta = mle.x
        alpha = 1 / (beta - 1)
        C = ((np.arange(len(word_counts)) + 1) ** (-alpha)).sum()
        params = {"alpha" : alpha, "C" : C}

    elif function == "third_function":
        mle = minimize_scalar(ll_third_function, bracket=(1 + 1e-10, 4), args=word_counts, method="brent")
        beta = mle.x
        alpha = 1 / (beta - 1)
        C = ((np.arange(len(word_counts)) + 1) ** (-alpha)).sum()
        params = {"alpha" : alpha, "C" : C}
    
    return params
    
def  ll_first_function(beta, count):
    return -np.sum(np.log(1 / (zeta(beta, 1) * (count ** beta))))

def ll_second_function(beta, counts):
    return -np.sum(np.log((1 / counts) ** (beta - 1) - (1 / (counts + 1)) ** (beta - 1)))

def ll_third_function(beta, count):
    likelihood = -np.sum(
        np.log(gamma(1) * (beta - 1) * gamma(count + 1 - beta) / (gamma(2 - beta) * gamma(count + 1)))
    )
    return likelihood
```

```python
params = mle_estimate(count, "first_function")
alpha_1 = params['alpha']
C_1 = params['C']

params = mle_estimate(count, "second_function")
alpha_2 = params['alpha']
C_2 = params['C']

params = mle_estimate(count, "third_function")
alpha_3 = params['alpha']
C_3 = params['C']
```

```python
fig = plt.figure(figsize = (10, 10))
plt.plot(rank, (count.sum() / C_1) * (rank ** (-alpha_1)), label='Best fit: First Function')
plt.plot(rank, (count.sum() / C_2) * (rank ** (-alpha_2)), label='Best fit: Second Function')
plt.plot(rank, (count.sum() / C_3) * (rank ** (-alpha_3)), label='Best fit: Third Function')
plt.scatter(rank, count, label='Data')
plt.title("Plot for Zipf's law", size = 15)
plt.xlabel("Rank of word (in log)", size = 15)
plt.ylabel("Frequency of occurance (in log)", size = 15)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import zipf, weibull_min, lognorm, kstest

dict_counts = np.log(count)

params_weibull = weibull_min.fit(dict_counts)
params_lognorm = lognorm.fit(dict_counts)

ll_zipf = zipf.logpmf(dict_counts, *params_zipf).sum()
ll_weibull = weibull_min.logpdf(dict_counts, *params_weibull).sum()
ll_lognorm = lognorm.logpdf(dict_counts, *params_lognorm).sum()

if ll_zipf > ll_weibull and ll_zipf > ll_lognorm:
    best_fit = 'Zipfian'
    best_params = params_zipf
elif ll_weibull > ll_lognorm:
    best_fit = 'Weibull'
    best_params = params_weibull
else:
    best_fit = 'Log-normal'
    best_params = params_lognorm


# Perform KS test to determine the goodness of fit
p_zipf = kstest(dict_counts, 'zipf', args=params_zipf)[1]
p_weibull = kstest(dict_counts, 'weibull_min', args=params_weibull)[1]
p_lognorm = kstest(dict_counts, 'lognorm', args=params_lognorm)[1]

# Print results
# print('Best fit:', best_fit)
# print('Best fit parameters:', best_params)
print('Zipf =', p_zipf, 'Weibull =', p_weibull, 'Log-normal =', p_lognorm)
```
